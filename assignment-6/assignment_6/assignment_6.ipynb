```python
##Answer to question 6.1.1
import scipy.io as sio
from sklearn.neighbors import KNeighborsClassifier
from statistics import variance
from toolbox.classification_plot import classification_plot
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


for i, (dataset, metric, n_neighbours) in enumerate(zip(["synth1", "synth2", "synth3", "synth4"], ['euclidean', 'seuclidean', 'seuclidean', 'euclidean'], [10, 5, 20, 20])):
    synth = sio.loadmat("./data/{}.mat".format(dataset))
    X_train = synth['X_train']
    y_train = synth['y_train'].ravel()
    X_test = synth['X_test']
    y_test = synth['y_test'].ravel()
    
    if metric == 'seuclidean':
        V = [variance(X_train[:, 0]), variance(X_train[:, 1])]
        knn = KNeighborsClassifier(n_neighbors=n_neighbours, metric=metric, metric_params={'V': V})
    else:
        knn = KNeighborsClassifier(n_neighbors=n_neighbours, metric=metric)
        
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    classification_plot(X_test, y_test, y_pred, X_train, y_train)
    plt.title("Figure {}: Clustering of dataset {} with distance metric {}"
              .format(i+1, dataset, metric))
    plt.show()
    
    cm = confusion_matrix(y_test, y_pred)
    plot_confusion_matrix(cm, y_test)
    print(classification_report(y_test, y_pred))
    
```

```python
##Answer to question 6.1.2
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import LeaveOneOut
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris_data = pd.read_excel("./data/iris.xls")
X = iris_data[['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']].to_numpy()
labels = iris_data['Type']
le = LabelEncoder()
y = le.fit_transform(labels)
loo = LeaveOneOut()
mean_errors = []
for k in range(1, 41):
    knn = KNeighborsClassifier(n_neighbors=k)
    errors = []
    for train_index, test_index in loo.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        knn.fit(X_train, y_train)
        y_pred = knn.predict(X_test)
        errors.append(y_pred != y_test)
    mean_errors.append(np.mean(errors))
plt.plot(range(1, 41), mean_errors)
plt.xlabel("Number of neighbors")
plt.ylabel("Mean error")
plt.title("Mean error as a function of the number of neighbors")
plt.show()
print("The optimal number of neighbors is {}".format(np.argmin(mean_errors)+1))
```

### 6.3.2 The kernel trick (1.25 points)
In order to solve the problem a linear kernel SVM has when facing such multimodal data, we can transform the data using a so called kernel. Several kernels exist, but in this case we'll look into a second degree polynomial kernel first.

There are several ways to construct a second degree polynomial kernel
- You can include or exclude the bias (the ^0 term)
- You can include or exclude non-interactions (X_1^d) terms

In this case, we can just add a third variable/feature to the data, namely the interaction term $x_1 * x_2$. **So we don't include the bias or the $x_1^2$ and $x_2^2$ terms.**

- Calculate the third interaction feature for the kernel. You can do this manually, or using the `PolynomialFeatures` function from Scikit-learn with the correct settings. Verify you end up with a $400*3$ matrix. Assign this NumPy array to a variable called `KX`.
- Make a 3D scatterplot of the transformed data, color the points by the label and add a legend
- Use a linear kernel SVM with otherwise default settings to predict `y`. What is the accuracy of the classifier now? Save the accuracy to a variable called `second_degree_svm_accuracy`. Does this match up with what you see in the plot? Can you find a separating hyperplane?

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVC

# Calculate the third interaction feature for the kernel
poly = PolynomialFeatures(degree=2, include_bias=False)
KX = poly.fit_transform(X)

# Make a 3D scatterplot of the transformed data
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(KX[:, 0], KX[:, 1], KX[:, 2], c=y)
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X1*X2")
plt.title("Scatter plot of the transformed Xor data")
plt.show()

# Use a linear kernel SVM to predict y
svc = SVC(kernel='linear')
svc.fit(KX, y)
second_degree_svm_accuracy = svc.score(KX, y)
print("The accuracy of the SVM is {}".format(second_degree_svm_accuracy))

# Rotate the plot to find a separating hyperplane
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(KX[:, 0], KX[:, 1], KX[:, 2], c=y)
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("X1*X2")
ax.view_init(azim=180)
plt.title("Scatter plot of the transformed Xor data")
plt.show()
```

The accuracy of the SVM is 0.9975, which matches up with what we see in the plot. We can find a separating hyperplane by rotating the plot. The hyperplane is the plane that separates the red and blue points.
